# First Attempt Code

```python
import cv2
import dlib
import numpy
from PIL import Image
import matplotlib.pyplot as plt
from matplotlib import path
%matplotlib inline

import sys

PREDICTOR_PATH = "../faceswap/shape_predictor_68_face_landmarks.dat"
SCALE_FACTOR = 1 
FEATHER_AMOUNT = 11

FACE_POINTS = list(range(17, 68))
MOUTH_POINTS = list(range(48, 61))
RIGHT_BROW_POINTS = list(range(17, 22))
LEFT_BROW_POINTS = list(range(22, 27))
RIGHT_EYE_POINTS = list(range(36, 42))
LEFT_EYE_POINTS = list(range(42, 48))
NOSE_POINTS = list(range(27, 36))
JAW_POINTS = list(range(0, 17))
MOUTH_INTERIOR_POINTS = list(range(61, 68))
ALL_POINTS = list(range(0, 68))

POINTS_FOR_HULL = [RIGHT_BROW_POINTS + LEFT_BROW_POINTS + JAW_POINTS]

# Points used to line up the images.
ALIGN_POINTS = (LEFT_BROW_POINTS + RIGHT_EYE_POINTS + LEFT_EYE_POINTS +
                               RIGHT_BROW_POINTS + NOSE_POINTS + MOUTH_POINTS)

# Amount of blur to use during colour correction, as a fraction of the
# pupillary distance.
COLOUR_CORRECT_BLUR_FRAC = 0.6

detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor(PREDICTOR_PATH)

class TooManyFaces(Exception):
    pass

class NoFaces(Exception):
    pass

def get_landmarks(im):
    rects = detector(im, 1)
    
    if len(rects) > 1:
        raise TooManyFaces
    if len(rects) == 0:
        raise NoFaces

    return numpy.matrix([[p.x, p.y] for p in predictor(im, rects[0]).parts()])

def draw_convex_hull(im, points, color):
    points = cv2.convexHull(points)
    cv2.fillConvexPoly(im, points, color=color)

def get_face_mask(im, landmarks):
    im = numpy.zeros(im.shape[:2], dtype=numpy.float64)

    for group in POINTS_FOR_HULL:
        draw_convex_hull(im,
                         landmarks[group],
                         color=1)

    im = numpy.array([im, im, im]).transpose((1, 2, 0))

    return im
    
def transformation_from_points(points1, points2):
    """
    Return an affine transformation [s * R | T] such that:

        sum ||s*R*p1,i + T - p2,i||^2

    is minimized.

    """
    # Solve the procrustes problem by subtracting centroids, scaling by the
    # standard deviation, and then using the SVD to calculate the rotation. See
    # the following for more details:
    #   https://en.wikipedia.org/wiki/Orthogonal_Procrustes_problem

    points1 = points1.astype(numpy.float64)
    points2 = points2.astype(numpy.float64)

    c1 = numpy.mean(points1, axis=0)
    c2 = numpy.mean(points2, axis=0)
    points1 -= c1
    points2 -= c2

    s1 = numpy.std(points1)
    s2 = numpy.std(points2)
    points1 /= s1
    points2 /= s2

    U, S, Vt = numpy.linalg.svd(points1.T * points2)

    # The R we seek is in fact the transpose of the one given by U * Vt. This
    # is because the above formulation assumes the matrix goes on the right
    # (with row vectors) where as our solution requires the matrix to be on the
    # left (with column vectors).
    R = (U * Vt).T

    return numpy.vstack([numpy.hstack(((s2 / s1) * R,
                                       c2.T - (s2 / s1) * R * c1.T)),
                         numpy.matrix([0., 0., 1.])])

def read_im_and_landmarks(fname):
    im = cv2.imread(fname, cv2.IMREAD_COLOR)
    im = cv2.resize(im, (im.shape[1] * SCALE_FACTOR,
                         im.shape[0] * SCALE_FACTOR))
    s = get_landmarks(im)

    return im, s

def warp_im(im, M, dshape):
    output_im = numpy.zeros(dshape, dtype=im.dtype)
    cv2.warpAffine(im,
                   M[:2],
                   (dshape[1], dshape[0]),
                   dst=output_im,
                   borderMode=cv2.BORDER_TRANSPARENT,
                   flags=cv2.WARP_INVERSE_MAP)
    return output_im
```

# Running the Code and Plotting Output

```python
# give paths to the images
paste_face = "images/catrina_makeup_1.jpg"
#paste_face = "face_mask_photoshop.png"
base_face = "images/grandma.jpg"

# read in the images and find the landmarks
im1, landmarks1 = read_im_and_landmarks(base_face)
im2, landmarks2 = read_im_and_landmarks(paste_face)

# determine how to distort face to fit on second
M = transformation_from_points(landmarks1[ALIGN_POINTS],
                               landmarks2[ALIGN_POINTS])

f, axarr = plt.subplots(2,4,figsize=(15,15))
axarr[0,3].imshow(im1)
axarr[0,3].set_title('im1')
axarr[1,3].imshow(im2)
axarr[1,3].set_title('im2')

# cut the paste face and add gaussian blur
mask = get_face_mask(im2, landmarks2)
axarr[0,0].imshow(mask)
axarr[0,0].set_title('mask')

# apply some transformation to base face
warped_mask = warp_im(mask, M, im1.shape)
axarr[0,1].imshow(warped_mask)
axarr[0,1].set_title('warped_mask')

# get the final mask we will put on base face
combined_mask = numpy.max([get_face_mask(im1, landmarks1), warped_mask],
                          axis=0)
axarr[0,2].imshow(combined_mask)
axarr[0,2].set_title('combined_mask')

# apply same transformation to 
warped_im2 = warp_im(im2, M, im1.shape)
axarr[1,0].imshow(warped_im2)
axarr[1,0].set_title('warped_im2')

final_mask = (warped_im2*combined_mask).astype('uint8')
axarr[1,1].imshow(final_mask)
axarr[1,1].set_title('final_mask')

# save final mask as png
final_mask_png = Image.fromarray(final_mask).convert("RGBA").save('final_mask.png')

fg = Image.fromarray(final_mask).convert('RGBA')
bg = Image.fromarray(im1).convert('RGBA')

output_im = Image.blend(bg, fg, 0.3)
output_im.save("output.png", "PNG")

# plot output
image = Image.open('output.png')
arr = numpy.asarray(image)
axarr[1,2].imshow(arr)
axarr[1,2].set_title('output')
# note I currently override what is there ... this box and warped_corrected_im2 both return weird values

plt.show()

plt.figure(figsize=(8,8))
image = Image.open('output.png')
arr = numpy.asarray(image)
plt.imshow(arr)
plt.show()
```

```python
from scipy.interpolate import interp1d
from pylab import *
from skimage import io, color
r,g,b = (207.,40.,57.)		#lipstick color

def inter(lx=[],ly=[],k1='quadratic'):
	unew = np.arange(lx[0], lx[-1]+1, 1)
	f2 = interp1d(lx, ly, kind=k1)
	return (f2,unew) # y-values, x-values

# putting points from dlib into same context as lipstick code
def get_inter(indices):
    return inter(np.floor(np.asarray(landmarkstest[indices][:,0])).flatten(), \
                  np.floor(np.asarray(landmarkstest[indices][:,1])).flatten())

# Code for the curves bounding the lips
o_u_l = get_inter(MOUTH_EXTERIOR_TOP_LEFT)
o_u_r = get_inter(MOUTH_EXTERIOR_TOP_RIGHT)
o_l = get_inter(MOUTH_EXTERIOR_BOTTOM)
i_u_l = get_inter(MOUTH_INTERIOR_TOP_LEFT)
i_u_r = get_inter(MOUTH_INTERIOR_TOP_RIGHT)
i_l = get_inter(MOUTH_INTERIOR_BOTTOM)

x = []	#will contain the x coordinates of points on lips
y = []  #will contain the y coordinates of points on lips

def ext(a,b,i): # a and b give range of x-values; i is the assigned y-value ...
	a,b = np.round(a),np.round(b)
	x.extend(arange(a,b,1).tolist())
	y.extend((ones(b-a)*i).tolist())
       
#print(o_u_l) # 1386-1978
#print(i_u_l) # 1386-1975
#print(o_l) # 1436-2549
#print(i_u_r) # 1975-2549
#print(o_u_r) # 1978-2549
#print(i_l) # 1436-2487

#print(int(o_u_l[1][0])) # 1386
#print(int(i_u_l[1][0])) # 1436

print(range(int(o_u_l[1][0]),int(i_u_l[1][0]+1)))
for i in range(int(o_u_l[1][0]),int(i_u_l[1][0]+1)):
	ext(o_u_l[0](i),o_l[0](i)+1,i)
print('1 successful')

print(range(int(i_u_l[1][0]),int(o_u_l[1][-1]+1)))
for i in range(int(i_u_l[1][0]),int(o_u_l[1][-1]+1)): # looping over i in range(1436, 1979)
	ext(o_u_l[0](i),i_u_l[0](i)+1,i)
	ext(i_l[0](i),o_l[0](i)+1,i)
print('2 successful')

print(range(int(i_u_r[1][-1]),int(o_u_r[1][-1]+1)))
for i in range(int(i_u_r[1][-1]),int(o_u_r[1][-1]+1)):
	ext(o_u_r[0](i),o_l[0](i)+1,i)
print('3 successful')

print(range(int(i_u_r[1][0]),int(i_u_r[1][-1]+1)))
for i in range(int(i_u_r[1][0]),int(i_u_r[1][-1]+1)):
	ext(o_u_r[0](i),i_u_r[0](i)+1,i)
	ext(i_l[0](i),o_l[0](i)+1,i)
print('4 successful')
    
val = color.rgb2lab((imtest[x,y]/255.).reshape(len(x),1,3)).reshape(len(x),3)
L,A,B = mean(val[:,0]),mean(val[:,1]),mean(val[:,2])
L1,A1,B1 = color.rgb2lab(np.array((r/255.,g/255.,b/255.)).reshape(1,1,3)).reshape(3,)
ll,aa,bb = L1-L,A1-A,B1-B
val[:,0] += ll
val[:,1] += aa
val[:,2] += bb

imtest, landmarkstest = read_im_and_landmarks("images/mouth_open_pic.jpg")
imtest[x,y] = color.lab2rgb(val.reshape(len(x),1,3)).reshape(len(x),3)*255
plt.gca().set_aspect('equal', adjustable='box')
plt.imshow(imtest)
plt.show()
```
